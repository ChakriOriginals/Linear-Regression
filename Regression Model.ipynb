```python
import pandas as pd
df = pd.read_csv("Food DataSet.csv")
df.head()
```

```python
import matplotlib.pyplot as matp
%matplotlib inline
```

```python
matp.scatter(df['No of Plates Wasted'],df['Food Ordered'])
```

```python
X = df[['No of Guests Invited']]
Y = df['Food Ordered']
```

```python
from sklearn.model_selection import train_test_split as tts
Xtrain, Xtest, Ytrain, Ytest = tts(X,Y,test_size=20/100) 
```

```python
Xtrain
```

```python
Xtest
```

```python
Ytest
```

```python
#Import the Linear Reg Model from sklearn lib
from sklearn.linear_model import LinearRegression 
clf = LinearRegression()
clf.fit(Xtrain, Ytrain)
```

```python
Xtest
```

```python
import numpy as np
Result = clf.predict(Xtest)
np_array = np. array(Result)
Final = np. around(np_array)
print(list(Final))
```

```python
Ytest
```

```python
clf.score(Xtest,Ytest)
```

```python
Guests = [input('Enter the number of Guests Invited: ')]
df1 = pd.DataFrame(Guests)
FinalResult = clf.predict(df1)
FinalResult = np. around(FinalResult)
print('The Optimal number of food to be ordered: %d ' %FinalResult)
```
